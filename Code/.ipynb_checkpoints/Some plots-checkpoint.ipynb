```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
import seaborn as sn
```

```python
data_path = "C:\\Users\\alicj\\Documents\\GitHub\\CSS_Alien\\Data\\"
data_1 = pd.read_csv(os.path.join(data_path, 'replicator_data_1.csv'))
data_2 = pd.read_csv(os.path.join(data_path, 'replicator_data_2_100032_iter.csv'))
```

```python
data_1.columns=["time step", "number of clusters", "number of clusters with tokens", "number of cells", "number of particles", "number of tokens", "total internal energy", "total kinetic energy"]
data_2.columns=["time step", "number of clusters", "number of clusters with tokens", "number of cells", "number of particles", "number of tokens", "total internal energy", "total kinetic energy"]
```

```python
fig, ax = plt.subplots(4, 2, figsize = (16, 18))
ax[0, 0].plot(data_1["time step"], data_1["number of clusters"], label = 'number of clusters')
ax[0, 0].plot(data_1["time step"], data_1["number of clusters with tokens"], label = 'number of clusters with tokens')
ax[0, 0].set_xlabel('Time step')
ax[0, 0].set_ylabel('Number of clusters')
ax[0, 0].set_title("data_1")
ax[0, 0].legend(loc = 5)

ax[0, 1].plot(data_2["time step"], data_2["number of clusters"], label = 'number of clusters')
ax[0, 1].plot(data_2["time step"], data_2["number of clusters with tokens"], label = 'number of clusters with tokens')
ax[0, 1].set_xlabel('Time step')
ax[0, 1].set_ylabel('Number of clusters')
ax[0, 1].set_title("data_2")
ax[0, 1].legend(loc = 5)

ax[1, 0].plot(data_1["time step"], data_1["number of clusters with tokens"])
ax[1, 0].set_xlabel('Time step')
ax[1, 0].set_ylabel('Number of clusters with tokens')

ax[1, 1].plot(data_2["time step"], data_2["number of clusters with tokens"])
ax[1, 1].set_xlabel('Time step')
ax[1, 1].set_ylabel('Number of clusters with tokens')


ax[1, 0].plot(data_1["time step"], data_1["number of clusters with tokens"])
ax[1, 0].set_xlabel('Time step')
ax[1, 0].set_ylabel('Number of clusters with tokens')

ax[1, 1].plot(data_2["time step"], data_2["number of clusters with tokens"])
ax[1, 1].set_xlabel('Time step')
ax[1, 1].set_ylabel('Number of clusters with tokens')


ax[2, 0].plot(data_1["total kinetic energy"], data_1["number of cells"])
ax[2, 0].set_xlabel('Total kinetic energy')
ax[2, 0].set_ylabel('Number of cells')

ax[2, 1].plot(data_2["total kinetic energy"], data_2["number of cells"])
ax[2, 1].set_xlabel('Total kinetic energy')
ax[2, 1].set_ylabel('Number of cells')


ax[3, 0].plot(data_1["time step"], data_1["total kinetic energy"])
ax[3, 0].set_xlabel('Time step')
ax[3, 0].set_ylabel('Total kinetic energy')

ax[3, 1].plot(data_2["time step"], data_2["total kinetic energy"])
ax[3, 1].set_xlabel('Time step')
ax[3, 1].set_ylabel('Total kinetic energy');
```

# Some comments:
- the plot of number of particles vs number of cells has a very weird shape for data_1 (also on log-log and semi-log axes)
- the plot of total kinetic energy vs number of cells on log-log axes kind of resembles some decay


# Correlations

```python
# for data_1
plt.figure(figsize=(12,9))
color = sn.diverging_palette(240, 240, as_cmap=True)
heatmap = sn.heatmap(data_1.corr(),annot=True, cmap=color, xticklabels=True, yticklabels=True) 
heatmap.set_yticklabels(heatmap.get_yticklabels(), rotation=30);
```

```python
# for data_2
plt.figure(figsize=(12,9))
color = sn.diverging_palette(240, 240, as_cmap=True)
heatmap = sn.heatmap(data_2.corr(),annot=True, cmap=color, xticklabels=True, yticklabels=True) 
heatmap.set_yticklabels(heatmap.get_yticklabels(), rotation=30);
```
